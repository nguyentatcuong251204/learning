```python
import matplotlib.pyplot as plt
from sklearn import tree
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score
from sklearn.model_selection import train_test_split
```

```python
iris_dataset=load_iris()
print(iris_dataset.feature_names)
model=DecisionTreeClassifier()
X_train,X_test,y_train,y_test=train_test_split(iris_dataset.data,iris_dataset.target,test_size=0.3,random_state=123)
model.fit(X_train,y_train)
y_predict=model.predict(X_test)
acc=accuracy_score(y_test,y_predict)
precision=precision_score(y_test,y_predict,average='macro')
recall=recall_score(y_test,y_predict,average='macro')
print(acc)
print(precision)
print(recall)

```

```python
plt.figure(figsize=(14,9))
tree.plot_tree(model,feature_names=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'],max_depth=5);
```
