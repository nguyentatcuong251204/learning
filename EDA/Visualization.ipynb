### MATPLOTLIB

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

<span style='color:blue'>PLOT

```python
arr1=np.array([1,3,5,7])
arr2=np.array([1,4,6,9])
plt.plot(arr1,arr2,marker='o')
plt.xlabel('x',size=20)
plt.ylabel('y',size=20)
plt.title("CORRELATION BETWEEN X AND Y")
plt.show()
```

```python
arr1=np.array([1,3,5,7])
arr2=np.array([1,4,6,9])
plt.plot(arr1,arr2,marker='o',ls='--') #CHANGE THE LINESTYLE
plt.xlabel('x',size=20)
plt.ylabel('y',size=20)
plt.title("CORRELATION BETWEEN X AND Y")
plt.show()
```

<span style='color:blue'>DISPLAY FUNCTION SINX,COSX

```python
X=np.linspace(0,10,1000)
plt.plot(X,np.sin(X),label='sinx')
plt.plot(X,np.cos(X),label='cosx')
plt.legend()
plt.xlabel("X",size=20)
plt.ylabel("Y",size=20)
plt.title("CORRELATION SINX AND COSX IN [0,10]")
```

```python
arr1=np.array([1,3,5,7])
arr2=np.array([1,4,6,9])
plt.plot(arr1,arr2,marker='o',ls='--',color='r') #CHANGE THE LINESTYLE
plt.xlabel('x',size=20)
plt.ylabel('y',size=20)
plt.title("CORRELATION BETWEEN X AND Y")
plt.show()
```

```python
arr1=np.array([1,3,5,7])
arr2=np.array([1,4,6,9])
plt.plot(arr1,arr2,marker='o',ls='--',lw=5,ms=20) #CHANGE THE LINEWWIDTH AND MARKER SIZE
plt.xlabel('x',size=20)
plt.ylabel('y',size=20)
plt.title("CORRELATION BETWEEN X AND Y")
plt.show()
```

```python
arr1=np.array([1,3,5,7])
arr2=np.array([1,4,6,9])
plt.plot(arr1,arr2,marker='o',ls='--',lw=5,ms=20) #CHANGE THE LINEWWIDTH
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.xlabel('x',size=20,fontdict=font2) #CHANGE THE FONT OF XLABEL
plt.ylabel('y',size=20,fontdict=font2) #CHANGE THE FONT OF YLABEL
plt.title("CORRELATION BETWEEN X AND Y",fontdict=font1)
plt.show()
```

<span style='color:blue'>PIE CHART

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

<span style='color:red'>*START AT ANGEL 90 DEGREE

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels = mylabels,startangle=90)
plt.show() 
```

<span style='color:red'>*EXPLODE

```python
y = np.array([35, 25, 25, 15])
labels = ["Apples", "Bananas", "Cherries", "Dates"]
explode=[0.2,0,0,0]
plt.pie(y, labels = labels,startangle=90,explode=explode,autopct='%1.1f%%')
plt.show() 
```

```python
fig,axs=plt.subplots(nrows=2,ncols=2)
arr1=np.array([1,3,5,7])
arr2=np.array([1,4,6,9])
axs[0][0].plot(arr1,arr2)
axs[0][1].plot(arr1,arr2,ls='--',color='r')
axs[1][0].plot(arr1,arr2,lw=5,ms=20,marker='o')
arr3=np.linspace(1,3,10)
arr4=np.linspace(3,5,10)
arr5=np.linspace(5,7,10)
arr_x=np.hstack((arr3,arr4))
arr_x=np.hstack((arr_x,arr5))
arr3=np.linspace(1,4,10)
arr4=np.linspace(4,6,10)
arr5=np.linspace(6,9,10)
arr_y=np.hstack((arr3,arr4))
arr_y=np.hstack((arr_y,arr5))
axs[1][1].scatter(arr_x,arr_y,c='r',s=3)
```

### SEABORN

```python
sns.get_dataset_names()
```

```python
tips_df=sns.load_dataset('tips')
tips_df.head()
```

<span style='color:blue'>HISTOGRAMPLOT

```python
sns.histplot(data=tips_df,x=tips_df['total_bill'])
```

<span style='color:blue'>KDE

```python
sns.kdeplot(tips_df['total_bill'])
```

<span style='color:blue'>DISTRIBUTION 

```python
sns.displot(data=tips_df,x='total_bill',kde=True)
```

```python
sns.displot(data=tips_df,x='total_bill',kde=True,col='sex')
```

<SPAN STYLE='color:blue'>BAR

```python
sns.barplot(data=tips_df,x='sex',y='tip',estimator=np.mean)
```

<SPAN STYLE='color:blue'>COUNTPLOT

```python
sns.countplot(data=tips_df,x=tips_df['sex'])
```

<span style='color:blue'>BOXPLOT

```python
sns.boxplot(data=tips_df,x='day',y='total_bill',hue='sex')
```

<span style='color:blue'>FACEGRID

```python
tips_fg=sns.FacetGrid(data=tips_df,col='time',row='smoker')
tips_fg.map(sns.scatterplot,'total_bill','tip')
```

```python
tips_fg=sns.FacetGrid(data=tips_df,col='time',hue='smoker')
tips_fg.map(sns.scatterplot,'total_bill','tip')
tips_fg.add_legend()
```
