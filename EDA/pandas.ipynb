```python
import numpy as np

def LU_decomposition(matrix):
    n = len(matrix)
    L = np.zeros((n, n))
    U = np.zeros((n, n))

    for j in range(n):
        L[j][j] = 1

        for i in range(j+1):
            s1 = sum(U[k][j] * L[i][k] for k in range(i))
            U[i][j] = matrix[i][j] - s1

        for i in range(j, n):
            s2 = sum(U[k][j] * L[i][k] for k in range(j))
            L[i][j] = (matrix[i][j] - s2) / U[j][j]

    return L, U

# Example usage
matrix = [[4, 3, -2], [2, 1, -3], [1, 3, -1]]
L, U = LU_decomposition(matrix)
print("MATRIX :")
print(np.array(matrix))
print("L:")
print(L)
print("U:")
print(U)
```
